**На вход функции подаётся словарь. Постройте из него серию и умножьте каждое значение серии на 2. Функция должна вернуть серию.**



```python
import pandas as pd

def solution(d):
    result = pd.Series(d) * 2
    return result
```

**На вход функции подаётся словарь. Постройте из него серию так чтобы индексы были расположены в порядке: e, a, c, b, d**

```python
import pandas as pd

def solution(d):
    result = pd.Series(d, index=['e', 'a', 'c', 'b', 'd'])
    return result
```

**На вход функции подаётся словарь. Постройте из него серию и добавьте еще один индекс g. Индексу g присвойте значение 10. Функция должна вернуть серию.**



```python
import pandas as pd

def solution(d):
    s1 = pd.Series(d)
    s1['g'] = 10
    return s1
```

**Проверьте, все ли значения серии заполнены? Верните одно значение False - если не все; True - если все заполнены**

```python
import pandas as pd

def solution(s1):
    result = s1.notna().all()
    return result
```
