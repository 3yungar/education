**Кстати, не обязательно что-то присваивать! Вы можете просто вытянуть из датафрейма эти значения на основе двумерной маски.**

```python
mask = df['col3'] > 9
mask
```

```python
df < 10
```

```python
df[df < 10] = 0
df
```

```python
mask = df < 10
mask
```

```python
mask.values
```

```python
df[~mask] = 100
df
```
