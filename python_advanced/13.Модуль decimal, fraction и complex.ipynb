Decimal —á–∏—Å–ª–∞, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞, —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–æ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π s. –î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª —Å—É–º–º—É –Ω–∞–∏–±–æ–ª—å—à–µ–≥–æ –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ Decimal —á–∏—Å–ª–∞.



```python
from decimal import Decimal as D

s = '''0.77 4.03 9.06 3.80 7.08 5.88 0.23 4.65 2.79 0.90 4.23 2.15 3.24 8.57 0.10 8.57 1.49 5.64
  3.63 8.36 1.56 6.67 1.46 5.26 4.83 7.23 1.22 1.02 7.82 9.97 5.40 9.79 9.82 2.78 2.96 0.07 1.72
  7.24 7.84 9.23 1.71 6.24 5.78 5.37 0.03 9.60 8.86 2.73 5.83 6.50'''

dlist = [D(x) for x in s.split()]
print(max(dlist) + min(dlist))
```

Decimal —á–∏—Å–ª–∞, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞, —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–æ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π s. –î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª –Ω–∞ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —Å—É–º–º—É –≤—Å–µ—Ö —á–∏—Å–µ–ª, –∞ –Ω–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ

```python
from decimal import Decimal as D

s = '''9.73 8.84 8.92 9.60 9.32 8.97 8.53 1.26 6.62 9.85 1.85 1.80 0.83 6.75 9.74
       9.11 9.14 5.03 5.03 1.34 3.52 8.09 7.89 8.24 8.23 5.22 0.30 2.59 1.25 6.24
       2.14 7.54 5.72 2.75 2.32 2.69 9.32 8.11 4.53 0.80 0.08 9.36 5.22 4.08 3.86
       5.56 1.43 8.36 6.29 5.13'''

dlist = [D(x) for x in s.split()]
print(sum(dlist))
print(*sorted(dlist, reverse=True)[:5])
```

–î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª —Å—É–º–º—É –Ω–∞–∏–±–æ–ª—å—à–µ–π –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–π —Ü–∏—Ñ—Ä—ã Decimal —á–∏—Å–ª–∞.

```python
from decimal import *
num = Decimal(input())

tup = num.as_tuple().digits
if num.quantize(Decimal(1), ROUND_DOWN) == 0:
    print(max(tup))
else:
    print(max(tup) + min(tup))
```

–î–µ—Å—è—Ç–∏—á–Ω—ã–µ —á–∏—Å–ª–∞ —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å–ø–∏—Å–∫–µ numbers –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫. –î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –¥–µ—Å—è—Ç–∏—á–Ω–æ–≥–æ —á–∏—Å–ª–∞ –≤—ã–≤–µ–ª –µ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤ –≤–∏–¥–µ –æ–±—ã–∫–Ω–æ–≤–µ–Ω–Ω–æ–π –¥—Ä–æ–±–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ:

–¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ = –æ–±—ã–∫–Ω–æ–≤–µ–Ω–Ω–∞—è –¥—Ä–æ–±—å

```python
from fractions import Fraction

numbers = ['6.34', '4.08', '3.04', '7.49', '4.45', '5.39',
           '7.82', '2.76', '0.71', '1.97', '2.54', '3.67',
           '0.14', '4.29', '1.84', '4.07', '7.26', '9.37',
           '8.11', '4.30', '7.16', '2.46', '1.27', '0.29',
           '5.12', '4.02', '6.95', '1.62', '2.26', '0.45',
           '6.91', '7.39', '0.52', '1.88', '8.38', '0.75',
           '0.32', '4.81', '3.31', '4.63', '7.84', '2.25',
           '1.10', '3.35', '2.05', '7.87', '2.40', '1.20',
           '2.58', '2.46']

for num in numbers:
    print(num , '=', Fraction(num))
```

–î–µ—Å—è—Ç–∏—á–Ω—ã–µ —á–∏—Å–ª–∞, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞, —Ö—Ä–∞–Ω—è—Ç—Å—è –≤ —Å—Ç—Ä–æ–∫–æ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π s. –î–æ–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π –∫–æ–¥, —á—Ç–æ–±—ã –æ–Ω –≤—ã–≤–µ–ª —Å—É–º–º—É –Ω–∞–∏–±–æ–ª—å—à–µ–≥–æ –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–≥–æ —á–∏—Å–ª–∞ –≤ –≤–∏–¥–µ –æ–±—ã–∫–Ω–æ–≤–µ–Ω–Ω–æ–π –¥—Ä–æ–±–∏.



```python
from fractions import Fraction

s = '0.78 4.3 9.6 3.88 7.08 5.88 0.23 4.65 2.79 0.90 4.23 2.15 3.24 8.57 0.10 8.57 1.49 5.64 3.63 8.36 1.56 6.67 1.46 5.26 4.83 7.13 1.22 1.02 7.82 9.97 5.40 9.79 9.82 2.78 2.96 0.07 1.72 7.24 7.84 9.23 1.71 6.24 5.78 5.37 0.03 9.60 8.86 2.73 5.83 6.50 0.123 0.00021'

numbers = [Fraction(num) for num in s.split()]
print(max(numbers) + min(numbers))
```

**–°–æ–∫—Ä–∞—Ç–∏—Ç–µ –¥—Ä–æ–±—å**

–î–∞–Ω—ã –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞
m –∏ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å–æ–∫—Ä–∞—â–∞–µ—Ç –¥—Ä–æ–±—å
m / n –∏ –≤—ã–≤–æ–¥–∏—Ç –µ–µ.

- –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞, —á–∏—Å–ª–∏—Ç–µ–ª—å –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å –¥—Ä–æ–±–∏, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

- –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

```python
from fractions import Fraction as F

m, n = int(input()), int(input())
print(F(m, n))
```

**–û–ø–µ—Ä–∞—Ü–∏–∏ –Ω–∞–¥ –¥—Ä–æ–±—è–º–∏**

–î–∞–Ω—ã –¥–≤–µ –¥—Ä–æ–±–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ a/b. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç –∏—Ö —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —á–∞—Å—Ç–Ω–æ–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ –Ω–µ–Ω—É–ª–µ–≤—ã–µ –¥—Ä–æ–±–∏, –∫–∞–∂–¥–∞—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å, –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —á–∞—Å—Ç–Ω–æ–µ –¥–≤—É—Ö –¥—Ä–æ–±–µ–π.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ç—Ä–µ—Ç–∏–π —Ç–µ—Å—Ç: –∏—Å—Ö–æ–¥–Ω—ã–µ –¥—Ä–æ–±–∏ —Å–æ–∫—Ä–∞—â–∞—Ç—å –Ω–µ –Ω—É–∂–Ω–æ, –∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω—É–∂–Ω–æ.

```python
from fractions import Fraction as F

s1, s2 = input(), input()
a1 = F(s1)
a2 = F(s2)

print(s1, '+', s2, '=' , a1 + a2)
print(s1, '-', s2, '=' , a1 - a2)
print(s1, '*', s2, '=' , a1 * a2)
print(s1, '/', s2, '=' , a1 / a2)
```

**–°—É–º–º–∞ –¥—Ä–æ–±–µ–π 1**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—é –≤—ã—Ä–∞–∂–µ–Ω–∏—è

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
n.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –†–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è –¥—Ä–æ–±—å –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º–æ–π.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ü–æ–¥—Ä–æ–±–Ω–µ–µ –æ —Ä—è–¥–µ –æ–±—Ä–∞—Ç–Ω—ã—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –º–æ–∂–Ω–æ –ø–æ—á–∏—Ç–∞—Ç—å —Ç—É—Ç.

```python
from fractions import Fraction as F

n = int(input())
total = 0

for i in range(1, n + 1):
    total += F(1, i**2)
print(total)
```

**–°—É–º–º–∞ –¥—Ä–æ–±–µ–π 2**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç –∏ –≤—ã–≤–æ–¥–∏—Ç —Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —Ä–∞–≤–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏—é –≤—ã—Ä–∞–∂–µ–Ω–∏—è

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
n.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –†–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∞—è –¥—Ä–æ–±—å –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ—Å–æ–∫—Ä–∞—Ç–∏–º–æ–π.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é factorial –∏–∑ –º–æ–¥—É–ª—è math.

```python
from fractions import Fraction as F
from math import factorial

n = int(input())
total = 0

for i in range(1, n + 1):
    total += F(1, factorial(i))

print(total)
```

- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í–æ–∑–º–æ–∂–Ω–æ –≤–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è gcd(), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å (–ù–û–î) –¥–≤—É—Ö —á–∏—Å–µ–ª. –§—É–Ω–∫—Ü–∏—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –≤ –º–æ–¥—É–ª–µ math.

```python
from math import gcd, ceil
from fractions import Fraction as F

n = int(input())

fraction_list = [(i, n - i) for i in range(1, ceil(n / 2)) if gcd(i, n - i) == 1]
a, b = fraction_list[0]
max_fraction = F(a, b)

for a, b in fraction_list[1:]:
    fract = F(a, b)
    if max_fraction < fract:
        max_fraction = fract
print(max_fraction)
```

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –Ω–∞ –∑–∞–¥–∞—á—É.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –í–æ–∑–º–æ–∂–Ω–æ –≤–∞–º –ø–æ—Ç—Ä–µ–±—É–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è gcd(), –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å (–ù–û–î) –¥–≤—É—Ö —á–∏—Å–µ–ª. –§—É–Ω–∫—Ü–∏—è —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –≤ –º–æ–¥—É–ª–µ math.

```python
from fractions import Fraction as F
from math import gcd

n = int(input())
result = []

for j in range(1, n + 1):
    for i in range(1, j):
        if gcd(i, j) == 1:
            result.append(F(i, j))

print(*sorted(result), sep='\n')
```
