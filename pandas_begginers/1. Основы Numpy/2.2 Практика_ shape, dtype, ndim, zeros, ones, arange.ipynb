**Создайте массив заполненный нулями у которого shape = (2, 5)**

```python
import numpy as np

def solution():
    arr = np.zeros((2, 5))
    return arr

solution()
```

**Создайте массив заполненный единицами у которого 5 строк и 8 столбцов (5 вложенных массивов длинной 8)**

```python
import numpy as np

def solution():
    arr = np.ones((5, 8))
    return arr

solution()
```

**Создайте одномерный numpy массив со значениями 0, 1, 2, .... 99 (диапазон значений от 0 до 99)**

```python
import numpy as np

def solution():
    arr = np.arange(100)
    return arr

solution()
```

**Создайте двумерный массив размером 5 х 2, который полностью будет заполнен значением 'G'**


```python
import numpy as np

def solution():
    arr = np.full((5, 2), 'G')
    return arr

solution()
```
