**Сравнение введенного числа с загаданным**

1. Организуйте сравнение введенного числа с загаданным числом:
- Если введенное число меньше загаданного числа, выведите текст: 'Ваше число меньше загаданного, попробуйте еще разок';
- Если введенное число больше загаданного числа, выведите текст: 'Ваше число больше загаданного, попробуйте еще разок';
- Если введенное число равно загаданному числу, выведите текст: 'Вы угадали, поздравляем!'.
2. Выведите прощальное сообщение пользователю: 'Спасибо, что играли в числовую угадайку. Еще увидимся...'.

**Улучшения проекта**
1. Добавьте подсчет попыток, сделанных пользователем. Когда число отгадано, программа должна показать количество попыток;
2. Добавьте возможность генерации нового числа (повторная игра), после того, как пользователь угадал число;
3. Добавить возможность указания правой границы для случайного выбора числа (от 1 до n).

```python
import random

def is_valid(num, right_board):
    if num.isdigit():
        return 0 < int(num) < right_board + 1
    return False

def check_input(right_board):
    while True:
        num = input()
        if is_valid(num, right_board):
            num = int(num)
            return num
        print(f'А может быть все-таки введем целое число от 1 до {right_board}')

def compare_number(num, x):
    if x > num:
        print('Ваше число меньше загаданного, попробуйте еще разок')
        return False
    if x < num:
        print('Ваше число больше загаданного, попробуйте еще разок')
        return False
    if x == num:
        print('Вы угадали, поздравляем!')
        print('Спасибо, что играли в числовую угадайку. Еще увидимся...')
        return True

def find_num(x, right_board):
    total = 0
    while True:
        total += 1
        num = check_input(right_board)
        if compare_number(num, x):
            return num, total
        


print('Добро пожаловать в числовую угадайку')
status = True
while status:
    print('Введите правую границу диапазона:', end=' ')
    right_board = int(input())
    x = random.randint(1, right_board)
    print('Ваше число:')
    res, total = find_num(x, right_board)
    print(f'Найденное число: {res}')
    print(f'Число попыток: {total}')
    print('Закончить игру? Напиши да. Хочешь продолжить игру отправь любой символ.')
    if input() == 'да':
        status = False
```

Напишите функцию generate_password(), которая принимает два аргумента:

и возвращает пароль.

Используя цикл for, сгенерируйте необходимое количество паролей.

```python
import random

def get_status(chars, status, txt):
    if status == 'да':
        chars += txt
    return chars

def get_status_strange(chars, status_strange_symbol):
    if status_strange_symbol == 'да':
        filter_chars = ''
        for x in chars:
            if x.lower() not in 'il1Lo0O':
                filter_chars += x
        return filter_chars
    
    return chars
    
def generate_password(length, chars):
    password = ''
    for i in range(length):
        password += chars[random.randrange(len(chars))]

    return password    
    

digits = '0123456789'
lowercase_letters = 'abcdefghijklmnopqrstuvwxyz'
uppercase_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
punctuation = '!#$%&*+-=?@^_.'
chars = ''

n = int(input('Введите количество паролей для генерации: '))
length = int(input('Введите длину одного пароля: '))
status_digit = input('Включать ли цифры 0123456789?: ')
status_upper = input('Включать ли прописные буквы ABCDEFGHIJKLMNOPQRSTUVWXYZ? ')
status_symbol = input('Включать ли символы !#$%&*+-=?@^_?: ')
status_strange_symbol = input('Исключать ли неоднозначные символы il1Lo0O?: ')

chars += lowercase_letters
chars = get_status(chars, status_digit, digits)
chars = get_status(chars, status_upper, uppercase_letters)
chars = get_status(chars, status_symbol, punctuation)
chars = get_status_strange(chars, status_strange_symbol)

passwords = [generate_password(length, chars) for _ in range(n)]
print(*passwords, sep='\n')
```

**Шифр Цезаря**

Описание проекта: требуется написать программу, способную шифровать и дешифровать текст в соответствии с алгоритмом Цезаря. Она должна запрашивать у пользователя следующие данные:

- направление: шифрование или дешифрование;
- язык алфавита: русский или английский;
- шаг сдвига (со сдвигом вправо).

Примечание 1. Считайте, что в русском языке 32 буквы (буква ё отсутствует).

Примечание 2. Неалфавитные символы — знаки препинания, пробелы, цифры — не меняются.

Примечание 3. Сохраните регистр символов. Например, текст: "Умом Россию не понять" при сдвиге на одну позицию вправо будет преобразован в: "Фнпн Спттйя ож рпоауэ".

Составляющие проекта:

- Целые числа (тип int);
- Модульная арифметика;
- Переменные;
- Ввод / вывод данных (функции input() и print());
- Условный оператор (if/elif/else);
- Цикл for/while;
- Строковые методы.

```python
eng_lower_alphabet = 'abcdefghijklmnopqrstuvwxyz'
eng_upper_alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
rus_lower_alphabet = "абвгдежзийклмнопрстуфхцчшщъыьэюя"
rus_upper_alphabet = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"

def get_code(txt, vector_coding, language, step):
    n = 32 if language == 'ru' else 26
    lower_alphabet = rus_lower_alphabet if language == 'ru' else eng_lower_alphabet
    upper_alphabet = rus_upper_alphabet if language == 'ru' else eng_upper_alphabet
    step = step if vector_coding == 'шифрование' else -step
    new_txt = ''

    for x in txt:
        if x.isalpha() and x.islower():
            indx = lower_alphabet.find(x)
            new_txt += lower_alphabet[(indx + step) % n]
        elif x.isalpha and x.isupper():
            indx = upper_alphabet.find(x)
            new_txt += upper_alphabet[(indx + step) % n]
        else:
            new_txt += x

    return new_txt



print('Выберите тип: шифрование или дешифрование. Введите 0 или 1 соответсвенно')
vector_coding = ['шифрование', 'дешифрование'][int(input())]
print('Выберите язык: английский или русский. Введите 0 или 1 соответсвенно')
language = ['eng', 'ru'][int(input())]
step = int(input('Укажите шаг сдвига: '))

print(f'Введите текст c режимом: {vector_coding} и языком: {language}')
txt = input()
res = get_code(txt, vector_coding, language, step)
print(res)
```

**Аве, Цезарь 🌶️**

На вход программе подается строка текста на английском языке, в которой нужно зашифровать все слова. Каждое слово строки следует зашифровать с помощью шифра Цезаря (циклического сдвига на длину этого слова). Строчные буквы при этом остаются строчными, а прописные – прописными.

Формат входных данных 
- На вход программе подается строка текста на английском языке.

Формат выходных данных
- Программа должна вывести зашифрованный текст в соответствии с условием задачи.

Примечание. Символы, не являющиеся английскими буквами, не изменяются.

```python
def get_code(txt):
    n = 26
    lower_alphabet = 'abcdefghijklmnopqrstuvwxyz'
    upper_alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    step = len(txt) % 26
    new_txt = ''

    for x in txt:
        if x.isalpha() and x.islower():
            indx = lower_alphabet.find(x)
            new_txt += lower_alphabet[(indx + step) % n]
        elif x.isalpha and x.isupper():
            indx = upper_alphabet.find(x)
            new_txt += upper_alphabet[(indx + step) % n]
        else:
            new_txt += x

    return new_txt

def get_list(txt):
    res = []
    word = ''
    for x in txt:
        if x.isalpha():
            word += x
        else:
            res.append(word)
            res.append(x)
            word = ''
    return res + [word]


txt = input()
words = get_list(txt)
res = [get_code(x) for x in words]
print(*res, sep='')
```

**Функция, возвращающая текущее состояние**

Функция display_hangman() принимает один аргумент tries – количество попыток угадывания слова и возвращает текущее состояние игры в графическом виде:

значение tries = 6 соответствует начальному положению, пустая висилица;

...

значение tries = 0 соответствует конечному положению, то есть проигрышу и 
полностью нарисованному телу повешенного. 

Примечание. Для вывода символа бэкслеша \ используется экранирование символа с помощью \, то есть комбинация \\.

```python
# функция получения текущего состояния
def display_hangman(tries):
    stages = [  # финальное состояние: голова, торс, обе руки, обе ноги
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                ''',
                # голова, торс, обе руки, одна нога
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                ''',
                # голова, торс, обе руки
                '''
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                ''',
                # голова, торс и одна рука
                '''
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                ''',
                # голова и торс
                '''
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                ''',
                # голова
                '''
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                ''',
                # начальное состояние
                '''
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                '''
    ]
    return stages[tries]
```

**Функция play()**

Напишите функцию play(), в которой будет осуществляться основная логика игры. Функция play() принимает в качестве аргумента слово word, сгенерированное функцией  get_word().

- отобразить начальное слово word_completion в виде строки с символом _ на каждую букву задуманного слова;

2. Необходимо обрабатывать ввод букв или слова целиком. Предусмотрите защиту от дурака, на случай если пользователь ввел символ, не являющийся буквой;
3. Если пользователь вводит уже названную букву или слово, то необходимо ему об этом сообщить, и не засчитывать попытку;
4. Если пользователь угадал букву, то требуется заменить все символы _ соответствующие этой букве;
5. Если пользователь угадал слово целиком, то следует его поздравить и вывести текст 'Поздравляем, вы угадали слово! Вы победили!';
6. Если пользователь исчерпал все свои попытки и не угадал слово, следует вывести загаданное слово.

Примечание. Переводите все символы в верхний регистр.

```python
def is_correct_input():
    while True:
        answer = input('Введите букву или слово целиком: ')
        if answer.isalpha():
            return answer

def check_use_again(answer, use_letter):
    return answer in use_letter

def check_answer(answer, word, word_completition):
    if answer == word:
        print('Поздравляем вы угадали слово! Вы победили!')
        return True
    
    indxs = list()
    update_word_completition = list()

    if answer in word:
        for i, x in enumerate(word):
            if x == answer:
                indxs.append(i)
    else:
        return False

    for i, x in enumerate(word_completition):
        if i in indxs:
            update_word_completition.append(answer)
            continue
        update_word_completition.append(x)

    return ''.join(update_word_completition)



def play(word):
    tries = 6
    word_completition = ''.join(['_' for _ in range(len(word))])
    use_letter = list()

    print('Давайте играть в угадайку слов!')

    while True:
        print('---' * 50)
        print(display_hangman(tries))
        print(word_completition)
        print('Осталось попыток:', tries)

        if tries == 0:
            print(15 * '-' +'ПОТРАЧЕНО' + 15 * '-')
            return word
    
        answer = is_correct_input()
        if check_use_again(answer, use_letter):
            print('Такая буква или слово уже были!')
            continue
        use_letter.append(answer)

        if check_answer(answer, word, word_completition) is True:
            return word

        if check_answer(answer, word, word_completition) is False:
            print('Мимо!')
            tries -= 1
            continue
        
        word_completition = check_answer(answer, word, word_completition)   
        print(word_completition)
        if word_completition == word:
            print('Поздравляем вы угадали слово! Вы победили!')
            return word_completition

```
