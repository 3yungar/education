**Выполните поэлементное сравнение двух столбцов:**

image.png



если значение в первом столбце больше чем значение во втором, тогда в результирующем массиве должно быть True, иначе False



```python
import numpy as np

def solution(arr):
    mask = arr[:, 0] > arr[:, 2]
    return mask
```
