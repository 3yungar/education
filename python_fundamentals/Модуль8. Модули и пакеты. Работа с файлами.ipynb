Подвиг 4. В программе имеется функция factorial (см. листинг). В начале программы (до определения функции) необходимо из модуля math импортировать функцию с тем же именем factorial, используя команду from, но так, чтобы они не "затирали" друг друга.

Уже объявленную функцию не менять. Выполнять функции не нужно, только прописать импорт.

```python
from math import factorial as fc

def factorial(n):
    p = 1
    for i in range(2, n+1):
        p *= i

    print("my factorial")
    return p

print(fc(5))
print(factorial(5))    
```

Подвиг 5. Из модуля random импортируйте только две функции: seed и randint. Затем, в программе выполните их, следующим образом:

    seed(1)
    print(randint(10, 50))

```python
from random import randint, seed

seed(1)
print(randint(10, 50))
```

Подвиг 6. Из модуля random импортируйте только две функции: seed и random, но у последней должен быть синоним rnd (имя, через которое она будет вызываться в программе). Выполните в программе эти функции, следующим образом:

    seed(10)
    print(round(rnd(), 2))

```python
from random import seed, random as rnd

seed(10)
print(round(rnd(), 2))
```

Подвиг 2. Имеется фрагмент программы (см. листинг ниже). При его выполнении возникает ошибка FileNotFoundError. Запишите конструкцию try / except, чтобы отображалось сообщение "File Not Found", если файл не удается открыть.



```python
try:
    f = open("abc.txt")
    r = f.read(1)
    f.close()
except FileNotFoundError:
    print('File Not Found')
```
