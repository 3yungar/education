**–ó–∞–≥–æ–ª–æ–≤–æ–∫ –ø—Ä–æ–≥—Ä–∞–º–º—ã**

1. –ü–æ–¥–∫–ª—é—á–∏—Ç–µ –º–æ–¥—É–ª—å random;
2. –°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ —Å–ª—É—á–∞–π–Ω–æ–µ —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ 100;
3. –í—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é: '–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ —á–∏—Å–ª–æ–≤—É—é —É–≥–∞–¥–∞–π–∫—É'.

**–§—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö**

–ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –º–æ–∂–µ—Ç –≤–≤–µ—Å—Ç–∏ –Ω–µ–≤–µ—Ä–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, –Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–µ —á–∏—Å–ª–æ, –∏–ª–∏ —á–∏—Å–ª–æ –ø—Ä–µ–≤—ã—à–∞—é—â–µ–µ 100. –í–∞–∂–Ω–æ –ø—Ä–µ–¥—É—Å–º–æ—Ç—Ä–µ—Ç—å —Ç–∞–∫—É—é –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å, —á—Ç–æ–±—ã –ø—Ä–æ–≥—Ä–∞–º–º–∞ –ø—Ä–æ–¥–æ–ª–∂–∞–ª–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Ä–∞–±–æ—Ç–∞—Ç—å. –û–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–∞–∫–æ–≥–æ —Ä–æ–¥–∞ —Å–∏—Ç—É–∞—Ü–∏–π –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –∑–∞—â–∏—Ç–æ–π –æ—Ç –¥—É—Ä–∞–∫–∞.

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é is_valid() –≤ –∫–æ—Ç–æ—Ä—É—é –ø–µ—Ä–µ–¥–∞–µ—Ç—Å—è –æ–¥–∏–Ω —Å—Ç—Ä–æ–∫–æ–≤—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç. –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ True –µ—Å–ª–∏ –ø–µ—Ä–µ–¥–∞–Ω–Ω—ã–π –∞—Ä–≥—É–º–µ–Ω—Ç —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º –æ—Ç 1 –¥–æ 100 –∏ False –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ. 



**–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ —Å –∑–∞–≥–∞–¥–∞–Ω–Ω—ã–º**

1. –û—Ä–≥–∞–Ω–∏–∑—É–π—Ç–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ —Å –∑–∞–≥–∞–¥–∞–Ω–Ω—ã–º —á–∏—Å–ª–æ–º:
- –ï—Å–ª–∏ –≤–≤–µ–¥–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞, –≤—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç: '–í–∞—à–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑–æ–∫';
- –ï—Å–ª–∏ –≤–≤–µ–¥–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞, –≤—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç: '–í–∞—à–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑–æ–∫';
- –ï—Å–ª–∏ –≤–≤–µ–¥–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Ä–∞–≤–Ω–æ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–º—É —á–∏—Å–ª—É, –≤—ã–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç: '–í—ã —É–≥–∞–¥–∞–ª–∏, –ø–æ–∑–¥—Ä–∞–≤–ª—è–µ–º!'.
2. –í—ã–≤–µ–¥–∏—Ç–µ –ø—Ä–æ—â–∞–ª—å–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é: '–°–ø–∞—Å–∏–±–æ, —á—Ç–æ –∏–≥—Ä–∞–ª–∏ –≤ —á–∏—Å–ª–æ–≤—É—é —É–≥–∞–¥–∞–π–∫—É. –ï—â–µ —É–≤–∏–¥–∏–º—Å—è...'.

**–£–ª—É—á—à–µ–Ω–∏—è –ø—Ä–æ–µ–∫—Ç–∞**
1. –î–æ–±–∞–≤—å—Ç–µ –ø–æ–¥—Å—á–µ—Ç –ø–æ–ø—ã—Ç–æ–∫, —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º. –ö–æ–≥–¥–∞ —á–∏—Å–ª–æ –æ—Ç–≥–∞–¥–∞–Ω–æ, –ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –ø–æ–∫–∞–∑–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ø—ã—Ç–æ–∫;
2. –î–æ–±–∞–≤—å—Ç–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –Ω–æ–≤–æ–≥–æ —á–∏—Å–ª–∞ (–ø–æ–≤—Ç–æ—Ä–Ω–∞—è –∏–≥—Ä–∞), –ø–æ—Å–ª–µ —Ç–æ–≥–æ, –∫–∞–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —É–≥–∞–¥–∞–ª —á–∏—Å–ª–æ;
3. –î–æ–±–∞–≤–∏—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å —É–∫–∞–∑–∞–Ω–∏—è –ø—Ä–∞–≤–æ–π –≥—Ä–∞–Ω–∏—Ü—ã –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ —á–∏—Å–ª–∞ (–æ—Ç 1 –¥–æ n).

```python
import random

def is_valid(num, right_board):
    if num.isdigit():
        return 0 < int(num) < right_board + 1
    return False

def check_input(right_board):
    while True:
        num = input()
        if is_valid(num, right_board):
            num = int(num)
            return num
        print(f'–ê –º–æ–∂–µ—Ç –±—ã—Ç—å –≤—Å–µ-—Ç–∞–∫–∏ –≤–≤–µ–¥–µ–º —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ –æ—Ç 1 –¥–æ {right_board}')

def compare_number(num, x):
    if x > num:
        print('–í–∞—à–µ —á–∏—Å–ª–æ –º–µ–Ω—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑–æ–∫')
        return False
    if x < num:
        print('–í–∞—à–µ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ –∑–∞–≥–∞–¥–∞–Ω–Ω–æ–≥–æ, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑–æ–∫')
        return False
    if x == num:
        print('–í—ã —É–≥–∞–¥–∞–ª–∏, –ø–æ–∑–¥—Ä–∞–≤–ª—è–µ–º!')
        print('–°–ø–∞—Å–∏–±–æ, —á—Ç–æ –∏–≥—Ä–∞–ª–∏ –≤ —á–∏—Å–ª–æ–≤—É—é —É–≥–∞–¥–∞–π–∫—É. –ï—â–µ —É–≤–∏–¥–∏–º—Å—è...')
        return True

def find_num(x, right_board):
    total = 0
    while True:
        total += 1
        num = check_input(right_board)
        if compare_number(num, x):
            return num, total
        


print('–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ —á–∏—Å–ª–æ–≤—É—é —É–≥–∞–¥–∞–π–∫—É')
status = True
while status:
    print('–í–≤–µ–¥–∏—Ç–µ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –¥–∏–∞–ø–∞–∑–æ–Ω–∞:', end=' ')
    right_board = int(input())
    x = random.randint(1, right_board)
    print('–í–∞—à–µ —á–∏—Å–ª–æ:')
    res, total = find_num(x, right_board)
    print(f'–ù–∞–π–¥–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ: {res}')
    print(f'–ß–∏—Å–ª–æ –ø–æ–ø—ã—Ç–æ–∫: {total}')
    print('–ó–∞–∫–æ–Ω—á–∏—Ç—å –∏–≥—Ä—É? –ù–∞–ø–∏—à–∏ –¥–∞. –•–æ—á–µ—à—å –ø—Ä–æ–¥–æ–ª–∂–∏—Ç—å –∏–≥—Ä—É –æ—Ç–ø—Ä–∞–≤—å –ª—é–±–æ–π —Å–∏–º–≤–æ–ª.')
    if input() == '–¥–∞':
        status = False
```

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –∑–∞–ø—Ä–∞—à–∏–≤–∞—Ç—å —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å–ª–µ–¥—É—é—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é:

1. –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–æ–ª–µ–π –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏;
2. –î–ª–∏–Ω—É –æ–¥–Ω–æ–≥–æ –ø–∞—Ä–æ–ª—è;
3. –í–∫–ª—é—á–∞—Ç—å –ª–∏ —Ü–∏—Ñ—Ä—ã 0123456789?
4. –í–∫–ª—é—á–∞—Ç—å –ª–∏ –ø—Ä–æ–ø–∏—Å–Ω—ã–µ –±—É–∫–≤—ã ABCDEFGHIJKLMNOPQRSTUVWXYZ?
5. –í–∫–ª—é—á–∞—Ç—å –ª–∏ —Å—Ç—Ä–æ—á–Ω—ã–µ –±—É–∫–≤—ã abcdefghijklmnopqrstuvwxyz?
6. –í–∫–ª—é—á–∞—Ç—å –ª–∏ —Å–∏–º–≤–æ–ª—ã !#$%&*+-=?@^_?
7. –ò—Å–∫–ª—é—á–∞—Ç—å –ª–∏ –Ω–µ–æ–¥–Ω–æ–∑–Ω–∞—á–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã il1Lo0O?

**–ù–∞—Å—Ç—Ä–æ–π–∫–∞ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º—ã—Ö –ø–∞—Ä–æ–ª–µ–π**

–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –≤–≤–µ–¥–µ–Ω–Ω–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏, —Å—Ñ–æ—Ä–º–∏—Ä—É–π—Ç–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é chars, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –≤—Å–µ —Å–∏–º–≤–æ–ª—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç –±—ã—Ç—å –≤ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º–æ–º –ø–∞—Ä–æ–ª–µ.

**–ì–µ–Ω–µ—Ä–∞—Ü–∏–∏ –ø–∞—Ä–æ–ª—è**

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é generate_password(), –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–≤–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞:

- length: –¥–ª–∏–Ω—É –ø–∞—Ä–æ–ª—è;
- chars: –∞–ª—Ñ–∞–≤–∏—Ç –∏–∑ —Å–∏–º–≤–æ–ª–æ–≤ –∫–æ—Ç–æ—Ä–æ–≥–æ —Å–æ—Å—Ç–æ–∏—Ç –ø–∞—Ä–æ–ª—å;
–∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ø–∞—Ä–æ–ª—å.

–ò—Å–ø–æ–ª—å–∑—É—è —Ü–∏–∫–ª for, —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–æ–ª–µ–π.

```python
import random

def get_status(chars, status, txt):
    if status == '–¥–∞':
        chars += txt
    return chars

def get_status_strange(chars, status_strange_symbol):
    if status_strange_symbol == '–¥–∞':
        filter_chars = ''
        for x in chars:
            if x.lower() not in 'il1Lo0O':
                filter_chars += x
        return filter_chars
    
    return chars
    
def generate_password(length, chars):
    password = ''
    for i in range(length):
        password += chars[random.randrange(len(chars))]

    return password    
    

digits = '0123456789'
lowercase_letters = 'abcdefghijklmnopqrstuvwxyz'
uppercase_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
punctuation = '!#$%&*+-=?@^_.'
chars = ''

n = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–æ–ª–µ–π –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏: '))
length = int(input('–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω—É –æ–¥–Ω–æ–≥–æ –ø–∞—Ä–æ–ª—è: '))
status_digit = input('–í–∫–ª—é—á–∞—Ç—å –ª–∏ —Ü–∏—Ñ—Ä—ã 0123456789?: ')
status_upper = input('–í–∫–ª—é—á–∞—Ç—å –ª–∏ –ø—Ä–æ–ø–∏—Å–Ω—ã–µ –±—É–∫–≤—ã ABCDEFGHIJKLMNOPQRSTUVWXYZ? ')
status_symbol = input('–í–∫–ª—é—á–∞—Ç—å –ª–∏ —Å–∏–º–≤–æ–ª—ã !#$%&*+-=?@^_?: ')
status_strange_symbol = input('–ò—Å–∫–ª—é—á–∞—Ç—å –ª–∏ –Ω–µ–æ–¥–Ω–æ–∑–Ω–∞—á–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã il1Lo0O?: ')

chars += lowercase_letters
chars = get_status(chars, status_digit, digits)
chars = get_status(chars, status_upper, uppercase_letters)
chars = get_status(chars, status_symbol, punctuation)
chars = get_status_strange(chars, status_strange_symbol)

passwords = [generate_password(length, chars) for _ in range(n)]
print(*passwords, sep='\n')
```

**–®–∏—Ñ—Ä –¶–µ–∑–∞—Ä—è**

–û–ø–∏—Å–∞–Ω–∏–µ –ø—Ä–æ–µ–∫—Ç–∞: —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å–ø–æ—Å–æ–±–Ω—É—é —à–∏—Ñ—Ä–æ–≤–∞—Ç—å –∏ –¥–µ—à–∏—Ñ—Ä–æ–≤–∞—Ç—å —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º –¶–µ–∑–∞—Ä—è. –û–Ω–∞ –¥–æ–ª–∂–Ω–∞ –∑–∞–ø—Ä–∞—à–∏–≤–∞—Ç—å —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å–ª–µ–¥—É—é—â–∏–µ –¥–∞–Ω–Ω—ã–µ:

- –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ: —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ –∏–ª–∏ –¥–µ—à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ;
- —è–∑—ã–∫ –∞–ª—Ñ–∞–≤–∏—Ç–∞: —Ä—É—Å—Å–∫–∏–π –∏–ª–∏ –∞–Ω–≥–ª–∏–π—Å–∫–∏–π;
- —à–∞–≥ —Å–¥–≤–∏–≥–∞ (—Å–æ —Å–¥–≤–∏–≥–æ–º –≤–ø—Ä–∞–≤–æ).

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –°—á–∏—Ç–∞–π—Ç–µ, —á—Ç–æ –≤ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ 32 –±—É–∫–≤—ã (–±—É–∫–≤–∞ —ë –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç).

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ù–µ–∞–ª—Ñ–∞–≤–∏—Ç–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã ‚Äî –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è, –ø—Ä–æ–±–µ–ª—ã, —Ü–∏—Ñ—Ä—ã ‚Äî –Ω–µ –º–µ–Ω—è—é—Ç—Å—è.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 3. –°–æ—Ö—Ä–∞–Ω–∏—Ç–µ —Ä–µ–≥–∏—Å—Ç—Ä —Å–∏–º–≤–æ–ª–æ–≤. –ù–∞–ø—Ä–∏–º–µ—Ä, —Ç–µ–∫—Å—Ç: "–£–º–æ–º –†–æ—Å—Å–∏—é –Ω–µ –ø–æ–Ω—è—Ç—å" –ø—Ä–∏ —Å–¥–≤–∏–≥–µ –Ω–∞ –æ–¥–Ω—É –ø–æ–∑–∏—Ü–∏—é –≤–ø—Ä–∞–≤–æ –±—É–¥–µ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω –≤: "–§–Ω–ø–Ω –°–ø—Ç—Ç–π—è –æ–∂ —Ä–ø–æ–∞—É—ç".

–°–æ—Å—Ç–∞–≤–ª—è—é—â–∏–µ –ø—Ä–æ–µ–∫—Ç–∞:

- –¶–µ–ª—ã–µ —á–∏—Å–ª–∞ (—Ç–∏–ø int);
- –ú–æ–¥—É–ª—å–Ω–∞—è –∞—Ä–∏—Ñ–º–µ—Ç–∏–∫–∞;
- –ü–µ—Ä–µ–º–µ–Ω–Ω—ã–µ;
- –í–≤–æ–¥ / –≤—ã–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö (—Ñ—É–Ω–∫—Ü–∏–∏ input() –∏ print());
- –£—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä (if/elif/else);
- –¶–∏–∫–ª for/while;
- –°—Ç—Ä–æ–∫–æ–≤—ã–µ –º–µ—Ç–æ–¥—ã.

```python
eng_lower_alphabet = 'abcdefghijklmnopqrstuvwxyz'
eng_upper_alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
rus_lower_alphabet = "–∞–±–≤–≥–¥–µ–∂–∑–∏–π–∫–ª–º–Ω–æ–ø—Ä—Å—Ç—É—Ñ—Ö—Ü—á—à—â—ä—ã—å—ç—é—è"
rus_upper_alphabet = "–ê–ë–í–ì–î–ï–ñ–ó–ò–ô–ö–õ–ú–ù–û–ü–†–°–¢–£–§–•–¶–ß–®–©–™–´–¨–≠–Æ–Ø"

def get_code(txt, vector_coding, language, step):
    n = 32 if language == 'ru' else 26
    lower_alphabet = rus_lower_alphabet if language == 'ru' else eng_lower_alphabet
    upper_alphabet = rus_upper_alphabet if language == 'ru' else eng_upper_alphabet
    step = step if vector_coding == '—à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ' else -step
    new_txt = ''

    for x in txt:
        if x.isalpha() and x.islower():
            indx = lower_alphabet.find(x)
            new_txt += lower_alphabet[(indx + step) % n]
        elif x.isalpha and x.isupper():
            indx = upper_alphabet.find(x)
            new_txt += upper_alphabet[(indx + step) % n]
        else:
            new_txt += x

    return new_txt



print('–í—ã–±–µ—Ä–∏—Ç–µ —Ç–∏–ø: —à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ –∏–ª–∏ –¥–µ—à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ. –í–≤–µ–¥–∏—Ç–µ 0 –∏–ª–∏ 1 —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–µ–Ω–Ω–æ')
vector_coding = ['—à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ', '–¥–µ—à–∏—Ñ—Ä–æ–≤–∞–Ω–∏–µ'][int(input())]
print('–í—ã–±–µ—Ä–∏—Ç–µ —è–∑—ã–∫: –∞–Ω–≥–ª–∏–π—Å–∫–∏–π –∏–ª–∏ —Ä—É—Å—Å–∫–∏–π. –í–≤–µ–¥–∏—Ç–µ 0 –∏–ª–∏ 1 —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–µ–Ω–Ω–æ')
language = ['eng', 'ru'][int(input())]
step = int(input('–£–∫–∞–∂–∏—Ç–µ —à–∞–≥ —Å–¥–≤–∏–≥–∞: '))

print(f'–í–≤–µ–¥–∏—Ç–µ —Ç–µ–∫—Å—Ç c —Ä–µ–∂–∏–º–æ–º: {vector_coding} –∏ —è–∑—ã–∫–æ–º: {language}')
txt = input()
res = get_code(txt, vector_coding, language, step)
print(res)
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º —è–∑—ã–∫–µ, –≤ –∫–æ—Ç–æ—Ä–æ–π –Ω—É–∂–Ω–æ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞—Ç—å –≤—Å–µ —Å–ª–æ–≤–∞. –ö–∞–∂–¥–æ–µ —Å–ª–æ–≤–æ —Å—Ç—Ä–æ–∫–∏ —Å–ª–µ–¥—É–µ—Ç –∑–∞—à–∏—Ñ—Ä–æ–≤–∞—Ç—å —Å –ø–æ–º–æ—â—å—é —à–∏—Ñ—Ä–∞ –¶–µ–∑–∞—Ä—è (—Ü–∏–∫–ª–∏—á–µ—Å–∫–æ–≥–æ —Å–¥–≤–∏–≥–∞ –Ω–∞ –¥–ª–∏–Ω—É —ç—Ç–æ–≥–æ —Å–ª–æ–≤–∞). –°—Ç—Ä–æ—á–Ω—ã–µ –±—É–∫–≤—ã –ø—Ä–∏ —ç—Ç–æ–º –æ—Å—Ç–∞—é—Ç—Å—è —Å—Ç—Ä–æ—á–Ω—ã–º–∏, –∞ –ø—Ä–æ–ø–∏—Å–Ω—ã–µ ‚Äì –ø—Ä–æ–ø–∏—Å–Ω—ã–º–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö 
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º —è–∑—ã–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –°–∏–º–≤–æ–ª—ã, –Ω–µ —è–≤–ª—è—é—â–∏–µ—Å—è –∞–Ω–≥–ª–∏–π—Å–∫–∏–º–∏ –±—É–∫–≤–∞–º–∏, –Ω–µ –∏–∑–º–µ–Ω—è—é—Ç—Å—è.

```python
def get_code(txt):
    n = 26
    lower_alphabet = 'abcdefghijklmnopqrstuvwxyz'
    upper_alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    step = len(txt) % 26
    new_txt = ''

    for x in txt:
        if x.isalpha() and x.islower():
            indx = lower_alphabet.find(x)
            new_txt += lower_alphabet[(indx + step) % n]
        elif x.isalpha and x.isupper():
            indx = upper_alphabet.find(x)
            new_txt += upper_alphabet[(indx + step) % n]
        else:
            new_txt += x

    return new_txt

def get_list(txt):
    res = []
    word = ''
    for x in txt:
        if x.isalpha():
            word += x
        else:
            res.append(word)
            res.append(x)
            word = ''
    return res + [word]


txt = input()
words = get_list(txt)
res = [get_code(x) for x in words]
print(*res, sep='')
```

```python
def is_correct_input():
    while True:
        answer = input('–í–≤–µ–¥–∏—Ç–µ –±—É–∫–≤—É –∏–ª–∏ —Å–ª–æ–≤–æ —Ü–µ–ª–∏–∫–æ–º: ')
        if answer.isalpha():
            return answer

def check_use_again(answer, use_letter):
    return answer in use_letter

def check_answer(answer, word, word_completition):
    if answer == word:
        print('–ü–æ–∑–¥—Ä–∞–≤–ª—è–µ–º –≤—ã —É–≥–∞–¥–∞–ª–∏ —Å–ª–æ–≤–æ! –í—ã –ø–æ–±–µ–¥–∏–ª–∏!')
        return True
    
    indxs = list()
    update_word_completition = list()

    if answer in word:
        for i, x in enumerate(word):
            if x == answer:
                indxs.append(i)
    else:
        return False

    for i, x in enumerate(word_completition):
        if i in indxs:
            update_word_completition.append(answer)
            continue
        update_word_completition.append(x)

    return ''.join(update_word_completition)



def play(word):
    tries = 6
    word_completition = ''.join(['_' for _ in range(len(word))])
    use_letter = list()

    print('–î–∞–≤–∞–π—Ç–µ –∏–≥—Ä–∞—Ç—å –≤ —É–≥–∞–¥–∞–π–∫—É —Å–ª–æ–≤!')

    while True:
        print('---' * 50)
        print(display_hangman(tries))
        print(word_completition)
        print('–û—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫:', tries)

        if tries == 0:
            print(15 * '-' +'–ü–û–¢–†–ê–ß–ï–ù–û' + 15 * '-')
            return word
    
        answer = is_correct_input()
        if check_use_again(answer, use_letter):
            print('–¢–∞–∫–∞—è –±—É–∫–≤–∞ –∏–ª–∏ —Å–ª–æ–≤–æ —É–∂–µ –±—ã–ª–∏!')
            continue
        use_letter.append(answer)

        if check_answer(answer, word, word_completition) is True:
            return word

        if check_answer(answer, word, word_completition) is False:
            print('–ú–∏–º–æ!')
            tries -= 1
            continue
        
        word_completition = check_answer(answer, word, word_completition)   
        print(word_completition)
        if word_completition == word:
            print('–ü–æ–∑–¥—Ä–∞–≤–ª—è–µ–º –≤—ã —É–≥–∞–¥–∞–ª–∏ —Å–ª–æ–≤–æ! –í—ã –ø–æ–±–µ–¥–∏–ª–∏!')
            return word_completition

```
