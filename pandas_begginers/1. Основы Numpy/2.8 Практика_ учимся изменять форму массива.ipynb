**На вход подается одномерный массив, преобразуйте его в матрицу с 5 строками и 6 столбцами**

```python
import numpy as np

def solution(arr):
    return arr.reshape(5, 6)
```

**На вход подается одномерный массив, преобразуйте его в трехмерный массив, где каждая матрица будет размером 4х5 и таких матриц должно быть 10**



```python
import numpy as np

def solution(arr):
    return arr.reshape(10, 4, 5)
```

**На вход подается двумерная матрица. Вытяните из нее последние три строки, но так чтобы строки были расположены в обратном порядке.**



```python
import numpy as np

def solution(arr):
    indx = np.arange(len(arr) - 3, len(arr))[::-1]
    return arr[indx]


arr = np.random.randint(0, 10, (4, 5))
print(arr)
print(solution(arr))
```

**На вход подается двумерная матрица. Вытяните из нее последние три строки, но так чтобы строки были расположены в обратном порядке. Затем возьмите только последние два столбца.**





```python
import numpy as np

def solution(arr):
    indx = np.arange(len(arr) - 3, len(arr))[::-1]
    return arr[indx][:, -2:]

arr = np.random.randint(0, 10, (4, 5))
print(arr)
print(solution(arr))
```

**На вход подается двумерная матрица размером 5х5. Преобразуйте ее в одномерный массив.**

```python
import numpy as np

def solution(arr):
    return arr.reshape(-1)

arr = np.random.randint(0, 10, (4, 5))
print(arr)
print(solution(arr))
```
