Подвиг 5. Вводится таблица целых чисел, записанных через пробел. Необходимо перемешать столбцы этой таблицы, используя функции shuffle и zip и вывести результат на экран (также в виде таблицы).

P. S. Для считывания списка целиком в программе уже записаны начальные строчки.

````python
import sys
import random

random.seed(1)

lst_in = list(map(str.strip, sys.stdin.readlines()))

lst_in = ['1 2 3 4', '5 6 7 8', '9 8 6 7']
z = zip(*(map(int, x.split()) for x in lst_in))
lst = []

for _ in range(len(lst_in[0].split())):
    lst.append(list(next(z)))

indx = list(range(len(lst)))
random.shuffle(indx)

for i in range(len(indx)):
    lst.append(lst[indx[i]])

z = zip(*lst[len(indx)::])
for _ in range(len(lst_in)):
    print(*next(z))
````

Значимый подвиг 7. Имеется двумерное игровое поле размером N x N (N - натуральное число, вводится с клавиатуры), представленное в виде вложенного списка:

    P = [[0] * N for i in range(N)]

Требуется расставить в нем случайным образом M = 10 единиц (целочисленных) так, чтобы они не соприкасались друг с другом (то есть, вокруг каждой единицы должны быть нули, либо граница поля). 

P.S. Поле на экран выводить не нужно (вообще ничего не нужно выводить), только сформировать.

````python
import random

random.seed(1)
N = int(input())
P = [[0] * N for i in range(N)]
M = 0

while M != 10:
    flag = False
    i_indx = random.randint(0, N - 1)
    j_indx = random.randint(0, N - 1)
    if P[i_indx][j_indx] == 1:
        flag = True
    else:    
        P[i_indx][j_indx] = 1

    for i in range(N - 1):
        for j in range(N - 1):
            if P[i][j] + P[i][j + 1] + P[i + 1][j] + P[i + 1][j + 1] > 1:
                flag = True
                P[i_indx][j_indx] = 0
                break
        if flag:
            break
    else:
        M += 1
````

````python
import numpy as np

total = 0
for i in range(N):
    total += np.sum(P[i]) 
    print(P[i])

print(total)    
````

# <h1>10.4 Конструкция match/case. Первое знакомство</h1>

Подвиг 5. Пользователь может ввести с клавиатуры следующие команды в виде строки:
```text
top или Top или TOP
bottom или Bottom или BOTTOM
right или Right или RIGHT
left или Left или LEFT
```
```python
cmd = input()
```
С помощью оператора match/case необходимо определить тип команды cmd и при совпадении вывести на экран сообщение в формате:

Команда <название команды малыми буквами>

Например, при вводе Top, должны на выходе получить:

Команда top

И так для всех четырех команд. Если тип команды не определен, то вывести строку:

Неверная команда
 

- Sample Input:

BOTTOM

- Sample Output:

Команда bottom

````python
cmd = input()

match cmd:
    case "top" | "Top" | "TOP":
        print('Команда top')
    case "bottom" | "Bottom" | "BOTTOM":
        print('Команда bottom')
    case "right" | "Right" | "RIGHT":
        print('Команда right')
    case "left" | "Left" | "LEFT":
        print('Команда left')
    case _:
        print('Неверная команда')
````

```python
    match value:
        # здесь продолжайте программу
```

Необходимо дописать оператор match/case со следующими шаблонами:

- если переменная value имеет тип int и больше нуля, то вернуть значение переменной value;
- если переменная value имеет тип float и находится в диапазоне [-100; 100], то вернуть значение переменной value;
- если переменная value имеет тип str, то просто вернуть ее значение.
- P. S. Вызывать функцию не нужно, только дописать шаблоны.

````python
def get_data(value):
    match value:
        case int() as value if value > 0:
            return value
        case float() as value if -100 <= value <= 100:
            return value
        case str() as value:
            return value
````

С помощью оператора match/case необходимо определить шаблоны для выделения следующей информации:

```text
автор, название
автор, название, цена
автор, название, год издания
автор, название, цена, год издания
```
Первый шаблон срабатывает, если есть только автор и название; второй, если появляется еще и цена; третий, если есть автор, название, год издания, но нет цены; последний, если имеются все данные.

````python
t = (int, str, str, float, int)
book = [t[i](x) if t[i] != str else x.strip() for i, x in enumerate(input().split(","))]

match book:
    case _, str() as author, str() as name:
        print('Yes')
    case _, str() as author, str() as name, float() as price:
        print('Yes')
    case _, str() as author, str() as name, int() as year:
        print('Yes')
    case _, str() as author, str() as name, float() as price, int() as year:
        print('Yes')
    case _, str() as author, str() as name, float() as price, int() as year, *_:
        print('Yes')
    case _:
        print('No')
````

- автор, название, год издания (год издания от 2020 и выше)
