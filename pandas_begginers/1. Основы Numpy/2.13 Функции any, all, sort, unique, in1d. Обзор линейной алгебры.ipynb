**На вход подаётся одномерный массив типа np.bool. Функция должна вернуть True если все значения в массиве равны False. Иначе функция должна вернуть False.**


```python
import numpy as np

def solution(m1d):
    return m1d.sum() == 0


m1d = np.random.choice(a=[False, True], size=10)
solution(m1d)
```

**На вход подаётся матрица. Посчитайте сумму элементов верхнего треугольника:**



```python
import numpy as np

def solution(m2d):
    return np.triu(m2d).sum()

m2d = np.random.randint(0, 10, (5, 5))
print(m2d)
solution(m2d)
```

**На вход подаётся одномерный массив. Посчитайте количество уникальных значений.**

```python
import numpy as np

def solution(m1d):
    return len(np.unique(m1d))

m1d = np.random.randint(0, 10, 10)
print(m1d)
solution(m1d)
```
