#Функции формирования массивов на основе данных

```python
a = np.array([(1, 2), (3, 4)])
a
```

```python
#Создание копии массива
b = np.copy(a)
b
```

```python
b[0] = 100
print(a, b, sep='\n')
```

```python
#x, y - в данном случае индексы(х - строка, у- столбец)
def get_range(x, y):
    return 100 * x + y


a = np.fromfunction(get_range, (2, 2))
a
```

```python
np.fromfunction(lambda x, y: x * 100 + y, (2, 2))
```

```python
#Формирование массива на основе любого итерируемого обьекта
np.fromiter('hello', dtype='U1')
```

```python
def getRange(N):
    for i in range(N):
        yield i


a = np.fromiter(getRange(4), dtype='int8')
a        
```

```python
#Формирование массива из строк
np.fromstring('1 2 3', dtype='int16', sep=' ')
```

```python
np.fromstring('1, 2, 3', dtype='int16', sep=',')
```
