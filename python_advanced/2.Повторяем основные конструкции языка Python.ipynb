**–ù–∞ easy**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ aa –∏ bb. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç:

- —Å—É–º–º—É —á–∏—Å–µ–ª aa –∏ bb;
- —Ä–∞–∑–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª aa –∏ bb;
- –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —á–∏—Å–µ–ª aa –∏ bb;
- —á–∞—Å—Ç–Ω–æ–µ —á–∏—Å–µ–ª aa –∏ bb;
- —Ü–µ–ª—É—é —á–∞—Å—Ç—å –æ—Ç –¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–∞ aa –Ω–∞ bb;
- –æ—Å—Ç–∞—Ç–æ–∫ –æ—Ç –¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–∞ aa –Ω–∞ bb;
- –∫–æ—Ä–µ–Ω—å –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∏–∑ —Å—É–º–º—ã –∏—Ö 10-—Ö —Å—Ç–µ–ø–µ–Ω–µ–π: \sqrt{a^{10} + b^{10}} 


–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ a –∏ b , (b!=0), –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
a, b = int(input()), int(input())
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print((a**10 + b**10)**0.5)
```

**–ò–Ω–¥–µ–∫—Å –º–∞—Å—Å—ã —Ç–µ–ª–∞**

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∏ –æ—Ü–µ–Ω–∫–∏ –∏–Ω–¥–µ–∫—Å–∞ –º–∞—Å—Å—ã —Ç–µ–ª–∞ (–ò–ú–¢) —á–µ–ª–æ–≤–µ–∫–∞. –ò–ú–¢ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Å–∏—Ç —á–µ–ª–æ–≤–µ–∫ –±–æ–ª—å—à–µ –∏–ª–∏ –º–µ–Ω—å—à–µ –Ω–æ—Ä–º—ã –¥–ª—è —Å–≤–æ–µ–≥–æ —Ä–æ—Å—Ç–∞. –ò–ú–¢ —á–µ–ª–æ–≤–µ–∫–∞ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

–ò–ú–¢= –º–∞—Å—Å–∞(–∫–≥) / (—Ä–æ—Å—Ç(–º) * —Ä–æ—Å—Ç(–º))

‚Äã–≥–¥–µ –º–∞—Å—Å–∞ –∏–∑–º–µ—Ä—è–µ—Ç—Å—è –≤ –∫–∏–ª–æ–≥—Ä–∞–º–º–∞—Ö, –∞ —Ä–æ—Å—Ç ‚Äî –≤ –º–µ—Ç—Ä–∞—Ö.

–ú–∞—Å—Å–∞ —á–µ–ª–æ–≤–µ–∫–∞ —Å—á–∏—Ç–∞–µ—Ç—Å—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π, –µ—Å–ª–∏ –µ–≥–æ –ò–ú–¢ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–µ–∂–¥—É 18.5 –∏ 25. –ï—Å–ª–∏ –ò–ú–¢ –º–µ–Ω—å—à–µ 18.5, —Ç–æ —Å—á–∏—Ç–∞–µ—Ç—Å—è, —á—Ç–æ —á–µ–ª–æ–≤–µ–∫ –≤–µ—Å–∏—Ç –Ω–∏–∂–µ –Ω–æ—Ä–º—ã. –ï—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ò–ú–¢ –±–æ–ª—å—à–µ 25, —Ç–æ —Å—á–∏—Ç–∞–µ—Ç—Å—è, —á—Ç–æ —á–µ–ª–æ–≤–µ–∫ –≤–µ—Å–∏—Ç –±–æ–ª—å—à–µ –Ω–æ—Ä–º—ã.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ "–û–ø—Ç–∏–º–∞–ª—å–Ω–∞—è –º–∞—Å—Å–∞", –µ—Å–ª–∏ –ò–ú–¢ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–µ–∂–¥—É 18.5 –∏ 25 (–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ). "–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞—è –º–∞—Å—Å–∞", –µ—Å–ª–∏ –ò–ú–¢ –º–µ–Ω—å—à–µ 18.5 –∏ "–ò–∑–±—ã—Ç–æ—á–Ω–∞—è –º–∞—Å—Å–∞", –µ—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ò–ú–¢ –±–æ–ª—å—à–µ 25.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞: –º–∞—Å—Å–∞ –∏ —Ä–æ—Å—Ç —á–µ–ª–æ–≤–µ–∫–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –í—Å–µ –≤—Ö–æ–¥–Ω—ã–µ —á–∏—Å–ª–∞ —è–≤–ª—è—é—Ç—Å—è –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–º–∏, –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –¥–ª—è –Ω–∏—Ö —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö float.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
m, l = float(input()), float(input())
IMT = m / l**2
if IMT < 18.5:
    print('–ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞—è –º–∞—Å—Å–∞')
elif 18.5 <= IMT <= 25:
    print('–û–ø—Ç–∏–º–∞–ª—å–Ω–∞—è –º–∞—Å—Å–∞')
else:
    print('–ò–∑–±—ã—Ç–æ—á–Ω–∞—è –º–∞—Å—Å–∞') 
```

**–°—Ç–æ–∏–º–æ—Å—Ç—å —Å—Ç—Ä–æ–∫–∏**

–î–∞–Ω–∞ —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ —Å—Ç—Ä–æ–∫–∏, –∏—Å—Ö–æ–¥—è –∏–∑ —Ç–æ–≥–æ, —á—Ç–æ –æ–¥–∏–Ω –ª—é–±–æ–π —Å–∏–º–≤–æ–ª (–≤ —Ç–æ–º —á–∏—Å–ª–µ –ø—Ä–æ–±–µ–ª) —Å—Ç–æ–∏—Ç 60 –∫–æ–ø–µ–µ–∫. –û—Ç–≤–µ—Ç –¥–∞–π—Ç–µ –≤ —Ä—É–±–ª—è—Ö –∏ –∫–æ–ø–µ–π–∫–∞—Ö –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –ø—Ä–∏–º–µ—Ä–∞–º–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å—Ç—Ä–æ–∫–∏.

```python
s = input()
cost = 60 * len(s)
print(f'{cost // 100} —Ä. {cost % 100} –∫–æ–ø.')
```

**–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤**

–î–∞–Ω–∞ —Å—Ç—Ä–æ–∫–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ —Å–ª–æ–≤, —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã—Ö –ø—Ä–æ–±–µ–ª–∞–º–∏. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ –≤ —ç—Ç–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –ö—Ä–æ–º–µ —Å–ª–æ–≤ –≤ —Ç–µ–∫—Å—Ç–µ –º–æ–≥—É—Ç –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –ø—Ä–æ–±–µ–ª—ã (–æ–¥–∏–Ω –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ).

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –†–µ—à–∏—Ç–µ –∑–∞–¥–∞—á—É –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É –∫–æ–¥–∞ üòé.

```python
len(input().split())
```

**–ó–æ–¥–∏–∞–∫**

–ö–∏—Ç–∞–π—Å–∫–∏–π –≥–æ—Ä–æ—Å–∫–æ–ø –Ω–∞–∑–Ω–∞—á–∞–µ—Ç –∂–∏–≤–æ—Ç–Ω—ã–º –≥–æ–¥—ã –≤ 12-–ª–µ—Ç–Ω–µ–º —Ü–∏–∫–ª–µ. –û–¥–∏–Ω 12-–ª–µ—Ç–Ω–∏–π —Ü–∏–∫–ª –ø–æ–∫–∞–∑–∞–Ω –≤ —Ç–∞–±–ª–∏—Ü–µ –Ω–∏–∂–µ. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, 2012 –≥–æ–¥ –±—É–¥–µ—Ç –æ—á–µ—Ä–µ–¥–Ω—ã–º –≥–æ–¥–æ–º –¥—Ä–∞–∫–æ–Ω–∞.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –≥–æ–¥ –∏ –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ —Å–≤—è–∑–∞–Ω–Ω–æ–≥–æ —Å –Ω–∏–º –∂–∏–≤–æ—Ç–Ω–æ–≥–æ. –í–∞—à–∞ –ø—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ —Ä–∞–±–æ—Ç–∞—Ç—å —Å –ª—é–±—ã–º –≥–æ–¥–æ–º, –Ω–µ —Ç–æ–ª—å–∫–æ —Ç–µ–º–∏, —á—Ç–æ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω—ã –≤ —Ç–∞–±–ª–∏—Ü–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –≥–æ–¥.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç ‚Äì –Ω–∞–∑–≤–∞–Ω–∏–µ –∂–∏–≤–æ—Ç–Ω–æ–≥–æ.

```python
name_of_years = ['–î—Ä–∞–∫–æ–Ω', '–ó–º–µ—è', '–õ–æ—à–∞–¥—å', '–û–≤—Ü–∞', '–û–±–µ–∑—å—è–Ω–∞', '–ü–µ—Ç—É—Ö', 
                 '–°–æ–±–∞–∫–∞', '–°–≤–∏–Ω—å—è', '–ö—Ä—ã—Å–∞', '–ë—ã–∫', '–¢–∏–≥—Ä', '–ó–∞—è—Ü']
year = int(input())
print(name_of_years[year % 12 - 8])
```

**–ü–µ—Ä–µ–≤–æ—Ä–æ—Ç —á–∏—Å–ª–∞**

–î–∞–Ω–æ –ø—è—Ç–∏–∑–Ω–∞—á–Ω–æ–µ –∏–ª–∏ —à–µ—Å—Ç–∏–∑–Ω–∞—á–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –∏–∑–º–µ–Ω–∏—Ç –ø–æ—Ä—è–¥–æ–∫ –µ–≥–æ –ø–æ—Å–ª–µ–¥–Ω–∏—Ö –ø—è—Ç–∏ —Ü–∏—Ñ—Ä –Ω–∞ –æ–±—Ä–∞—Ç–Ω—ã–π.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ –ø—è—Ç–∏–∑–Ω–∞—á–Ω–æ–µ –∏–ª–∏ —à–µ—Å—Ç–∏–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø–æ–ª—É—á–∏—Ç—Å—è –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —Ä–∞–∑–≤–æ—Ä–æ—Ç–∞, —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –≤ —É—Å–ª–æ–≤–∏–∏ –∑–∞–¥–∞—á–∏. –ß–∏—Å–ª–æ –Ω—É–∂–Ω–æ –≤—ã–≤–æ–¥–∏—Ç—å –±–µ–∑ –Ω–µ–∑–Ω–∞—á–∞—â–∏—Ö –Ω—É–ª–µ–π.

```python
n = input()
print(int(n[::-1])) if len(n) == 5 else print(int(n[0] + n[:-6:-1]))
```

**Standard American Convention**

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—Å—Ç–∞–≤–ª—è–µ—Ç –≤ –∑–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∑–∞–ø—è—Ç—ã–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∏–º —Å–æ–≥–ª–∞—à–µ–Ω–∏–µ–º –æ –∑–∞–ø—è—Ç—ã—Ö –≤ –±–æ–ª—å—à–∏—Ö —á–∏—Å–ª–∞—Ö.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n, \, (0 < n < 10^{100})

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ —Å –∑–∞–ø—è—Ç—ã–º–∏ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
n = input()[::-1]
standart = ''
for i in range(0, len(n), 3):
    if i == 0:
        standart += n[i:i + 3]
        continue
    standart += ',' + n[i:i + 3]
print(standart[::-1])
```

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫. –ö–∞–∂–¥–∞—è —Å–ª–µ–¥—É—é—â–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –¥–≤—É—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ‚Äî –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ç–æ—á–∫–∏ (—Å–Ω–∞—á–∞–ª–∞ –∞–±—Å—Ü–∏—Å—Å–∞ ‚Äì x, –∑–∞—Ç–µ–º –æ—Ä–¥–∏–Ω–∞—Ç–∞ ‚Äì y), —Ä–∞–∑–¥–µ–ª–µ–Ω–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–º –ø—Ä–æ–±–µ–ª–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫, –ª–µ–∂–∞—â–∏—Ö –≤ –∫–∞–∂–¥–æ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω–æ–π —á–µ—Ç–≤–µ—Ä—Ç–∏, –∫–∞–∫ –≤ –ø—Ä–∏–º–µ—Ä–∞—Ö.

```python
n = int(input())
count1, count2, count3, count4 = 0, 0, 0, 0

for _ in range(n):
    x, y = map(int, input().split())
    if x > 0 and y > 0:
        count1 += 1
    elif x < 0 and y > 0:
        count2 += 1
    elif x < 0 and y < 0:
        count3 += 1
    elif x > 0 and y < 0:
        count4 += 1

print('–ü–µ—Ä–≤–∞—è —á–µ—Ç–≤–µ—Ä—Ç—å:', count1)
print('–í—Ç–æ—Ä–∞—è —á–µ—Ç–≤–µ—Ä—Ç—å:', count2)
print('–¢—Ä–µ—Ç—å—è —á–µ—Ç–≤–µ—Ä—Ç—å:', count3)
print('–ß–µ—Ç–≤–µ—Ä—Ç–∞—è —á–µ—Ç–≤–µ—Ä—Ç—å:', count4)
```

```python
lst = list(map(int, input().split()))
count = 0

for i in range(1, len(lst)):
    if lst[i] > lst[i - 1]:
        count += 1
print(count)
```

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∏–∑–º–µ–Ω–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫, —Ä–∞–∑–¥–µ–ª—è—è –µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç—ã –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º.

```python
lst = list(map(int, input().split()))

for i in range(0, len(lst) - 1, 2):
    a, b = lst[i], lst[i + 1]
    lst[i], lst[i + 1] = b, a

print(*lst)
```

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–∏—Å–ª–æ n(0<n<1000) ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª –≤ –Ω–∞–±–æ—Ä–µ. –í –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö nn —Å—Ç—Ä–æ–∫–∞—Ö –≤–≤–æ–¥—è—Ç—Å—è —Ü–µ–ª—ã–µ —á–∏—Å–ª–∞, —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏–µ –Ω–∞–±–æ—Ä (–º–æ–≥—É—Ç –ø–æ–≤—Ç–æ—Ä—è—Ç—å—Å—è). –ó–∞—Ç–µ–º —Å–ª–µ–¥—É–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –∫–æ—Ç–æ—Ä–æ–µ —è–≤–ª—è–µ—Ç—Å—è –∏–ª–∏ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ–º –¥–≤—É—Ö –∫–∞–∫–∏—Ö-—Ç–æ —á–∏—Å–µ–ª –∏–∑ –Ω–∞–±–æ—Ä–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–î–ê¬ª –∏–ª–∏ ¬´–ù–ï–¢¬ª –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –°–∞–º–æ –Ω–∞ —Å–µ–±—è —á–∏—Å–ª–æ –∏–∑ –Ω–∞–±–æ—Ä–∞ —É–º–Ω–æ–∂–∏—Ç—å—Å—è –Ω–µ –º–æ–∂–µ—Ç, –¥—Ä—É–≥–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, –¥–≤–∞ –º–Ω–æ–∂–∏—Ç–µ–ª—è –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å —Ä–∞–∑–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã –≤ –Ω–∞–±–æ—Ä–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –î–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã.

```python
n = int(input())
lst = [int(input()) for _ in range(n)]
total = int(input())
flag = False

for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
          if lst[i] * lst[j] == total:
              flag = True
              print('–î–ê')
              break
    if flag:
        break
else:
    print('–ù–ï–¢')
```

- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –±—É–∫–≤ —Ä—É—Å—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞ "–û" –∏ "–†".

- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–¥—Ä—è–¥ –≤—ã–ø–∞–≤—à–∏—Ö –†–µ—à–µ–∫.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ –≤—ã–ø–∞–≤—à–∏—Ö –†–µ—à–µ–∫ –Ω–µ—Ç, —Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ 00.

```python
s = input()
while '–û–û' in s:
    s = s.replace('–û–û', '–û')

lst = s.split('–û')
max = len(lst[0])

for x in lst:
    if max < len(x):
        max = len(x)
print(max)
```

- –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–∏—Å–ª–æ n ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–æ–≤. –í –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö n —Å—Ç—Ä–æ–∫–∞—Ö –≤–≤–æ–¥—è—Ç—Å—è —Å—Ç—Ä–æ–∫–∏, —Å–æ–¥–µ—Ä–∂–∞—â–∏–µ –ª–∞—Ç–∏–Ω—Å–∫–∏–µ —Å—Ç—Ä–æ—á–Ω—ã–µ –±—É–∫–≤—ã –∏ —Ü–∏—Ñ—Ä—ã, –≤ –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –æ—Ç 5 –¥–æ 100 —Å–∏–º–≤–æ–ª–æ–≤.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–æ–º–µ—Ä–∞ –∑–∞—Ä–∞–∂–µ–Ω–Ω—ã—Ö —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–æ–≤ —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —Ö–æ–ª–æ–¥–∏–ª—å–Ω–∏–∫–æ–≤ –Ω–µ—Ç, –Ω–∏—á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç—å –Ω–µ –Ω—É–∂–Ω–æ.

```python
n = int(input())
lst = [input() for _ in range(n)]

for i, s in enumerate(lst):
    flag = True
    for letter in 'anton':
        if letter not in s:
            flag = False
            break
        s = s[s.find(letter):]
    if flag:
        print(i + 1, end=' ')
```

- –§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ —Å–ª–æ–≤–æ, –∑–∞–ø–∏—Å–∞–Ω–Ω–æ–µ —Å—Ç—Ä–æ—á–Ω—ã–º–∏ —Ä—É—Å—Å–∫–∏–º–∏ –±—É–∫–≤–∞–º–∏ –±–µ–∑ –±—É–∫–≤—ã "—ë".

- –§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º —Å—Ç—Ä–æ–∫–∏, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞–≤–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ä–∞–∑–Ω—ã—Ö –±—É–∫–≤ –≤ —Å—Ç—Ä–æ–∫–µ, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∞–µ—Ç—Å—è –ø—É—Ç–µ–º –∫–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏–∏ –≤–≤–µ–¥–µ–Ω–Ω–æ–≥–æ —Å–ª–æ–≤–∞ –∏ —Å—Ç—Ä–æ–∫–∏ "–∑–∞–ø—Ä–µ—Ç–∏–ª –±—É–∫–≤—É".

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –¢–µ–∫—Å—Ç –∏—Å—Ö–æ–¥–Ω–æ–π –ø–µ—Å–Ω–∏ –≤ –ø–µ—Ä–≤–æ–º —Ç–µ—Å—Ç–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ü–æ–µ–º –∏ —Ä–µ—à–∞–µ–º, –¥—Ä—É–∑—å—è, –ø–æ–µ–º –∏ —Ä–µ—à–∞–µ–º üòÇ.

```python
b = ['–∞', '–±', '–≤', '–≥', '–¥', '–µ', '–∂', '–∑', '–∏', '–π', '–∫', '–ª', '–º', 
     '–Ω', '–æ', '–ø', '—Ä', '—Å', '—Ç', '—É', '—Ñ', '—Ö', '—Ü', '—á', '—à', '—â',
     '—ä', '—ã', '—å', '—ç', '—é', '—è']

s = input().lower() + ' –∑–∞–ø—Ä–µ—Ç–∏–ª –±—É–∫–≤—É'

for x in b:
    if x not in s or len(x) == 0:
        continue
    s += ' ' + x
    print(s)
    while x in s or '  ' in s:
        s = s.replace(x, '').strip()
        s = s.replace('  ', ' ')
```
