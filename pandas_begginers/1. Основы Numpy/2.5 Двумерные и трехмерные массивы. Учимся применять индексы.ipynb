#**Маски**

Маска это массив (одномерный или многомерный), который состоит из булевых значений. Если True - то элемент будет вытянут, если False - нет.

```python
m = np.array([5, 3, 6, 8, 10])
m
```

```python
mask = m > 5
m > 5
```

```python
m[mask]
```

```python
m[m > 5]
```

```python
m = np.arange(5)
mask = [True, True, False, False, True]
m
```

```python
m2 = m[mask]
m2
```

```python
m2[0] = 1
m2
```

```python
m
```

**помните, что при обращении по маске значения будут скопированы**

```python
visits[0], visits[-1]
```

```python
users == 'Андрей'
```
