**На вход подается матрица. Посчитайте минимальные значения по каждому столбцу**

```python
import numpy as np

def solution(m2d):
    result = m2d.min(axis=0)
    return result

m2d = np.random.randint(0, 10, (3, 5))
print(m2d)
solution(m2d)
```

**На вход подается матрица. Посчитайте максимальные значения по каждой строке**

```python
import numpy as np

def solution(m2d):
    result = m2d.max(axis=1)
    return result

m2d = np.random.randint(0, 10, (3, 5))
print(m2d)
solution(m2d)
```

**Ранее пройденная задача, но попробуем решить ее снова:**

На вход подается одномерный массив в котором хранится информация о количестве картофеля на каждом складе (всего 4 склада). Посчитайте для каждого склада, сколько процентов картофеля от общего количества хранится на каждом складе.

Общее количество картофеля на всех складах (определите сами через функцию).

Не забудьте что каждое значение надо умножить на 100.

Округлять значения не требуется! Знак процента тоже добавлять не следует. Просто выполните арифметические операции.

```python
import numpy as np

def solution(arr):
    result = arr / arr.sum() * 100
    return result


arr = np.random.randint(200, 400, 4)
print(arr)
solution(arr)
```

**Вам передан одномерный массив. Если значение меньше или равно нулю, то замените его на нуль. Если значение больше нуля, то замените на 1.**



```python
import numpy as np

def solution(arr):
    res = np.where(arr <= 0, 0, 1)
    return res

arr = np.random.randint(-10, 10, 10)
print(arr)
print(solution(arr))
```

**Вам передана матрица. Если значение меньше нуля, то возьмите его по модулю. Если значение больше нуля, то извлеките из него квадратный корень.**



```python
import numpy as np

def solution(m2d):
    res = np.where(m2d < 0, np.abs(m2d), np.sqrt(m2d))
    return res


m2d = np.random.randint(-9, 10, (4, 5))
print(m2d)
print(solution(m2d))
```

**На вход подается матрица. Для каждого столбца определите индекс строки в которой содержится минимальное значение.**

```python
import numpy as np

def solution(m2d):
    res = m2d.argmin(axis=0)
    return res


m2d = np.random.randint(0, 100, (4, 5))
print(m2d)
solution(m2d)
```

**На вход подается матрица. Для каждой строки определите индекс столбца в котором содержится максимальное значение.**

```python
import numpy as np

def solution(m2d):
    return m2d.argmax(axis=1)


m2d = np.random.randint(0, 10, (4, 5))
print(m2d)
solution(m2d)
```
