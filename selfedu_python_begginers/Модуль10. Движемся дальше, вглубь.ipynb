Подвиг 7. Вводится зашифрованное слово. Шифрование кодов символов этого слова было проведено с помощью битовой операции XOR с ключом key=123. То есть, каждый символ был преобразован по алгоритму:

x = ord(x) ^ key

Здесь ord - функция, возвращающая код символа x. Расшифруйте введенное слово и выведите его на экран.

P. S. Подсказка: для преобразования кода в символ используйте функцию chr.

```python
s = input()
key = 123
lst = [chr(ord(x) ^ key) for x in s]
print(''.join(lst))
```

Подвиг 5. Вводится таблица целых чисел, записанных через пробел. Необходимо перемешать столбцы этой таблицы, используя функции shuffle и zip и вывести результат на экран (также в виде таблицы).

P. S. Для считывания списка целиком в программе уже записаны начальные строчки.

```python
import sys
import random

random.seed(1)

lst_in = list(map(str.strip, sys.stdin.readlines()))

lst_in = ['1 2 3 4', '5 6 7 8', '9 8 6 7']
z = zip(*(map(int, x.split()) for x in lst_in))
lst = []

for _ in range(len(lst_in[0].split())):
    lst.append(list(next(z)))

indx = list(range(len(lst)))
random.shuffle(indx)

for i in range(len(indx)):
    lst.append(lst[indx[i]])

z = zip(*lst[len(indx)::])
for _ in range(len(lst_in)):
    print(*next(z))
```

Подвиг 6. Вводятся имена студентов в одну строчку через пробел. Требуется случайным образом выбрать трех студентов из этого списка, используя функцию sample. (Полагается, что в исходном списке более трех студентов). Результат вывести на экран в одну строчку через пробел.



```python
import random

random.seed(1)
names = input().split()
a = random.sample(names, 3)
print(*a)
```

Значимый подвиг 7. Имеется двумерное игровое поле размером N x N (N - натуральное число, вводится с клавиатуры), представленное в виде вложенного списка:

    P = [[0] * N for i in range(N)]

Требуется расставить в нем случайным образом M = 10 единиц (целочисленных) так, чтобы они не соприкасались друг с другом (то есть, вокруг каждой единицы должны быть нули, либо граница поля). 

P.S. Поле на экран выводить не нужно (вообще ничего не нужно выводить), только сформировать.

```python
import random

random.seed(1)
N = int(input())
P = [[0] * N for i in range(N)]
M = 0

while M != 10:
    flag = False
    i_indx = random.randint(0, N - 1)
    j_indx = random.randint(0, N - 1)
    if P[i_indx][j_indx] == 1:
        flag = True
    else:    
        P[i_indx][j_indx] = 1

    for i in range(N - 1):
        for j in range(N - 1):
            if P[i][j] + P[i][j + 1] + P[i + 1][j] + P[i + 1][j + 1] > 1:
                flag = True
                P[i_indx][j_indx] = 0
                break
        if flag:
            break
    else:
        M += 1
```

```python
import numpy as np

total = 0
for i in range(N):
    total += np.sum(P[i]) 
    print(P[i])

print(total)    
```
