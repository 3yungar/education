**–ù–∞–ø–æ–ª–Ω–∏—Ç–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ set1 —Å–æ–¥–µ—Ä–∂–∏–º—ã–º —Ç–∞–∫, —á—Ç–æ–±—ã –ø—Ä–æ–≥—Ä–∞–º–º–∞ –≤—ã–≤–µ–ª–∞ {'p'}.**

```python
set1 = {'a', 't', 'f', 'p'}
set2 = {'a', 't', 'f'}

print(set1 - set2)
```

**–î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ**

–£—á–∏—Ç–µ–ª—å –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ –≤ –∫–ª–∞—Å—Å–µ –∏ –ø–æ–ª—É—á–∏–ª —Å–ª–µ–¥—É—é—â–∏–µ –æ—Ç–≤–µ—Ç—ã: –∏–∑ 
n —à–∫–æ–ª—å–Ω–∏–∫–æ–≤ —É 
m –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ —Å—ä–µ–ª–∞ —Å–æ–±–∞–∫–∞, —É 
k –æ—Ç–∫–ª—é—á–∏–ª–∏ —Å–≤–µ—Ç, –∞ 
p —É—á–µ–Ω–∏–∫–æ–≤ –ø–æ—Å—Ç–∏–≥–ª–∏ –æ–±–∞ –Ω–µ—Å—á–∞—Å—Ç—å—è. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å–∫–æ–ª—å–∫–æ —á–µ–ª–æ–≤–µ–∫ –≤—ã–ø–æ–ª–Ω–∏–ª–æ –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —á–∏—Å–ª–∞ 
n,m,k,p, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, —Å–¥–µ–ª–∞–≤—à–∏—Ö –¥–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ.

```python
n = int(input())
m = int(input())
k = int(input())
p = int(input())

print(n - (m + k - p))
```

**–í–æ—Å—Ö–æ–¥**

–ù–∞ —Å–ø—É—Ç–Ω–∏–∫–µ ¬´–í–æ—Å—Ö–æ–¥¬ª —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –ø—Ä–∏–±–æ—Ä –¥–ª—è –∏–∑–º–µ—Ä–µ–Ω–∏—è —Å–æ–ª–Ω–µ—á–Ω–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏. –ö–∞–∂–¥—É—é –º–∏–Ω—É—Ç—É –æ–Ω –ø–µ—Ä–µ–¥–∞—ë—Ç –≤ –æ–±—Å–µ—Ä–≤–∞—Ç–æ—Ä–∏—é –ø–æ –∫–∞–Ω–∞–ª—É —Å–≤—è–∑–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–Ω–µ—Ä–≥–∏–∏ —Å–æ–ª–Ω–µ—á–Ω–æ–≥–æ –∏–∑–ª—É—á–µ–Ω–∏—è. –î–ª—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –Ω–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ –¥–µ—Ä–∂–∞—Ç—å –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –¥–∞–Ω–Ω—ã–µ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫–∞–∑–∞–Ω–∏–π —Å–ø—É—Ç–Ω–∏–∫–∞, –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞, —Å–æ–¥–µ—Ä–∂–∞—â–∞—è —á–∏—Å–ª–∞ ‚Äì –ø–æ–∫–∞–∑–∞–Ω–∏—è —Å–ø—É—Ç–Ω–∏–∫–∞ ¬´–í–æ—Å—Ö–æ–¥¬ª. –ß–∏—Å–ª–∞ —É–∫–∞–∑—ã–≤–∞—é—Ç—Å—è —á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª –∏ –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤–µ–¥—É—â–∏—Ö –Ω—É–ª–µ–π.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫–∞–∑–∞–Ω–∏–π, –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã—Ö –∞–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –±—É–¥–µ—Ç –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω –≤–µ—Ä–Ω–æ.

```python
lst = [int(x) for x in input().split()]
myset = set(lst)
print(len(lst) - len(myset))
```

**–ì–æ—Ä–æ–¥–∞**

–¢–∏–º—É—Ä –∏ –†—É—Å–ª–∞–Ω –∏–≥—Ä–∞—é—Ç –≤ –∏–≥—Ä—É –≥–æ—Ä–æ–¥–∞. –û–Ω–∏ –æ—á–µ–Ω—å –ª—é–±—è—Ç —ç—Ç—É –∏–≥—Ä—É –∏ –∑–Ω–∞—é—Ç –º–Ω–æ–≥–æ –≥–æ—Ä–æ–¥–æ–≤, –æ—Å–æ–±–µ–Ω–Ω–æ –¢–∏–º—É—Ä, –æ–¥–Ω–∞–∫–æ –∫ –∫–æ–Ω—Ü—É –∏–≥—Ä—ã –≤–≤–∏–¥—É —Å–≤–æ–µ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç–∞ –∑–∞–±—ã–≤–∞—é—Ç, –∫–∞–∫–∏–µ –≥–æ—Ä–æ–¥–∞ —É–∂–µ –Ω–∞–∑—ã–≤–∞–ª–∏.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, —Å—á–∏—Ç—ã–≤–∞—é—â—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± –∏–≥—Ä–µ –∏ —Å–æ–æ–±—â–∞—é—â—É—é —Ä–µ–±—è—Ç–∞–º, —á—Ç–æ –æ—á–µ—Ä–µ–¥–Ω–æ–π –≥–æ—Ä–æ–¥ –Ω–∞–∑–≤–∞–Ω –ø–æ–≤—Ç–æ—Ä–Ω–æ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ 
n ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–∑–≤–∞–Ω–Ω—ã—Ö –≥–æ—Ä–æ–¥–æ–≤, –≤ –ø–æ—Å–ª–µ–¥—É—é—â–∏—Ö 
n —Å—Ç—Ä–æ–∫–∞—Ö –≤–≤–æ–¥—è—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–Ω—ã–µ –≥–æ—Ä–æ–¥–∞ –∏ –µ—â—ë –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ —Å –Ω–æ–≤—ã–º, —Ç–æ–ª—å–∫–æ —á—Ç–æ –Ω–∞–∑–≤–∞–Ω—ã–º –≥–æ—Ä–æ–¥–æ–º.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ OK, –µ—Å–ª–∏ —ç—Ç–æ—Ç –≥–æ—Ä–æ–¥ –µ—â—ë –Ω–µ –≤—Å–ø–æ–º–∏–Ω–∞–ª–∏, –∏ REPEAT, –µ—Å–ª–∏ –≥–æ—Ä–æ–¥ —É–∂–µ –±—ã–ª –Ω–∞–∑–≤–∞–Ω.

```python
n = int(input())
cities = set()

for _ in range(n):
    cities.add(input())

print('REPEAT') if input() in cities else print('OK')
```

**–ö–Ω–∏–≥–∏ –Ω–∞ –ø—Ä–æ—á—Ç–µ–Ω–∏–µ**

–†—É—Å–ª–∞–Ω –ø–æ–ª—É—á–∏–ª –≤ –∫–æ–Ω—Ü–µ —É—á–µ–±–Ω–æ–≥–æ –≥–æ–¥–∞ —Å–ø–∏—Å–æ–∫ –ª–∏—Ç–µ—Ä–∞—Ç—É—Ä—ã –Ω–∞ –ª–µ—Ç–æ. –¢–µ–ø–µ—Ä—å –µ–º—É –Ω–∞–¥–æ –≤—ã—è—Å–Ω–∏—Ç—å, –∫–∞–∫–∏–µ –∫–Ω–∏–≥–∏ –∏–∑ —ç—Ç–æ–≥–æ —Å–ø–∏—Å–∫–∞ —É –Ω–µ–≥–æ –µ—Å—Ç—å. –£ –†—É—Å–ª–∞–Ω–∞ –Ω–∞ –∫–æ–º–ø—å—é—Ç–µ—Ä–µ –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º —Ñ–∞–π–ª–µ –∑–∞–ø–∏—Å–∞–Ω—ã –≤—Å–µ –∫–Ω–∏–≥–∏ –∏–∑ –µ–≥–æ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –¥–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏ –∏–∑ —Å–ø–∏—Å–∫–∞ –Ω–∞ –ø—Ä–æ—á—Ç–µ–Ω–∏–µ, –µ—Å—Ç—å –æ–Ω–∞ —É –†—É—Å–ª–∞–Ω–∞ –∏–ª–∏ –Ω–µ—Ç.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ 
m ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–Ω–∏–≥ –≤ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–µ –†—É—Å–ª–∞–Ω–∞. –í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ 
n ‚Äî  –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–Ω–∏–≥ –≤ —Å–ø–∏—Å–∫–µ –Ω–∞ –ª–µ—Ç–æ. –î–∞–ª–µ–µ –∏–¥—É—Ç 
m —Å—Ç—Ä–æ–∫ —Å –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –∫–Ω–∏–≥ –∏–∑ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∏ 
n —Å—Ç—Ä–æ–∫ –Ω–∞–∑–≤–∞–Ω–∏–π –∏–∑ —Å–ø–∏—Å–∫–∞ –Ω–∞ –ª–µ—Ç–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ 
n —Å—Ç—Ä–æ–∫, –≤ –∫–∞–∂–¥–æ–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–∞–ø–∏—Å–∞–Ω–æ —Å–ª–æ–≤–æ YES, –µ—Å–ª–∏ –∫–Ω–∏–≥–∞ –Ω–∞–π–¥–µ–Ω–∞ –≤ –±–∏–±–ª–∏–æ—Ç–µ–∫–µ, –∏ NO, –µ—Å–ª–∏ –Ω–µ—Ç.

```python
m = int(input())
n = int(input())

available_books = {input() for _ in range(m)}
required_books = [input() for _ in range(n)]

for book in required_books:
    if book in available_books:
        print('YES')
        continue
    print('NO')
```

**–°—Ç—Ä–∞–Ω–Ω–æ–µ —É–≤–ª–µ—á–µ–Ω–∏–µ**

–ö–∞–∫ –∏–∑–≤–µ—Å—Ç–Ω–æ, –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏ —Å—Ç—Ä–∞–Ω–Ω—ã–µ –ª—é–¥–∏. –ù–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –∏—Å–∫–ª—é—á–µ–Ω–∏—è –∏ –¢–∏–º—É—Ä ‚Äî –∞–≤—Ç–æ—Ä –¥–∞–Ω–Ω–æ–≥–æ –∫—É—Ä—Å–∞. –ö–∞–∂–¥—ã–π –¥–µ–Ω—å –¢–∏–º—É—Ä —Ä–µ—à–∞–µ—Ç —Ä–æ–≤–Ω–æ –¥–≤–µ —Å–ª–æ–∂–Ω—ã–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–µ –∑–∞–¥–∞—á–∏. –†–µ—à–∞—è –ø–µ—Ä–≤—É—é –∑–∞–¥–∞—á—É, –æ–Ω –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç –Ω–∞ –ø–µ—Ä–≤–æ–º –ª–∏—Å—Ç–æ—á–∫–µ –≤—Å–µ —á–∏—Å–ª–∞, –∫–æ—Ç–æ—Ä—ã–µ –≤ –Ω–µ–π –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è. –î–∞–ª–µ–µ –æ–Ω –¥–µ–ª–∞–µ—Ç –ø–∞—É–∑—É –∏ –±–µ—Ä–µ—Ç—Å—è –∑–∞ –≤—Ç–æ—Ä—É—é –∑–∞–¥–∞—á—É. –ó–∞—Ç–µ–º –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç –Ω–∞ –≤—Ç–æ—Ä–æ–º –ª–∏—Å—Ç–æ—á–∫–µ –≤—Å–µ —á–∏—Å–ª–∞, –∫–æ—Ç–æ—Ä—ã–µ –≤ –Ω–µ–π –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –æ–Ω –±–µ—Ä–µ—Ç –µ—â–µ –æ–¥–∏–Ω –ª–∏—Å—Ç–æ–∫ –∏ –≤—ã–ø–∏—Å—ã–≤–∞–µ—Ç –Ω–∞ –Ω–µ–≥–æ –≤—Å–µ —Å–æ–≤–ø–∞–¥–∞—é—â–∏–µ —á–∏—Å–ª–∞ –∏–∑ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö –ª–∏—Å—Ç–æ—á–∫–æ–≤. –ï—Å–ª–∏ —Ç–∞–∫–∏–µ —á–∏—Å–ª–∞ –µ—Å—Ç—å, –¥–µ–Ω—å —É–¥–∞–ª—Å—è, –µ—Å–ª–∏ –æ–±—â–∏—Ö —á–∏—Å–µ–ª –Ω–µ—Ç, –¢–∏–º—É—Ä —Å—á–∏—Ç–∞–µ—Ç –¥–µ–Ω—å –Ω–µ—É–¥–∞—á–Ω—ã–º.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –æ–±—â–∏–µ —á–∏—Å–ª–∞ –¥–≤—É—Ö –ª–∏—Å—Ç–æ—á–∫–æ–≤ –∏–ª–∏ —Å–æ–æ–±—â–∞–µ—Ç, —á—Ç–æ –¥–µ–Ω—å –Ω–µ —É–¥–∞–ª—Å—è üòè

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å —á–∏—Å–ª–∞–º–∏: –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —á–∏—Å–ª–∞ —Å –ø–µ—Ä–≤–æ–≥–æ –ª–∏—Å—Ç–æ—á–∫–∞, –≤–æ –≤—Ç–æ—Ä–æ–π —Å–æ –≤—Ç–æ—Ä–æ–≥–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —á–∏—Å–ª–∞, –≤—Å—Ç—Ä–µ—Ç–∏–≤—à–∏–µ—Å—è –Ω–∞ –æ–±–æ–∏—Ö –ª–∏—Å—Ç–∫–∞—Ö –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø–æ—Ä—è–¥–∫–µ, –ª–∏–±–æ —Å–ª–æ–≤–æ—Å–æ—á–µ—Ç–∞–Ω–∏–µ BAD DAY, –µ—Å–ª–∏ —Ç–∞–∫–∏—Ö —á–∏—Å–µ–ª –Ω–µ—Ç.

```python
first_sheet = {int(x) for x in input().split()}
second_sheet = {int(x) for x in input().split()}

result = first_sheet.intersection(second_sheet)
if len(result) == 0:
    print('BAD DAY')
else:
    print(*sorted(result, reverse=True))
```

**–û–Ω–ª–∞–π–Ω-—à–∫–æ–ª–∞ BEEGEEK 1**

–ü—Ä–∏ –ø—Ä–∏—ë–º–µ –Ω–æ–≤—ã—Ö —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –≤ –æ–Ω–ª–∞–π–Ω-—à–∫–æ–ª—É BEEGEEK –µ—ë —Ä—É–∫–æ–≤–æ–¥–∏—Ç–µ–ª—å —Ç–µ—Å—Ç–∏—Ä—É–µ—Ç –Ω–µ —Ç–æ–ª—å–∫–æ –ø—Ä–æ—Ñ–µ—Å—Å–∏–æ–Ω–∞–ª—å–Ω—ã–µ –∫–∞—á–µ—Å—Ç–≤–∞ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞, –Ω–æ –∏ –µ–≥–æ –ø–∞–º—è—Ç—å.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å —á–∏—Å–ª–∞–º–∏: –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –ø–æ–∫–∞–∑–∞–Ω–Ω—ã–µ –∫–∞–Ω–¥–∏–¥–∞—Ç—É, –∞ –≤–æ –≤—Ç–æ—Ä–æ–π –æ—Ç–≤–µ—Ç—ã –∫–∞–Ω–¥–∏–¥–∞—Ç–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ YES, –µ—Å–ª–∏ –∫–∞–Ω–¥–∏–¥–∞—Ç –ø—Ä–æ—à–µ–ª –∏—Å–ø—ã—Ç–∞–Ω–∏–µ –∏ –µ–≥–æ –º–æ–∂–Ω–æ –±—Ä–∞—Ç—å –Ω–∞ —Ä–∞–±–æ—Ç—É –∏ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
test = {x for x in input().split()}
answer = {x for x in input().split()}

print('YES') if len(test.intersection(answer)) == len(test) == len(answer) else print('NO')
```

- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∏–∑—É—á–∞—é—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —É—á–µ–Ω–∏–∫–æ–≤ –Ω–µ –æ–∫–∞–∂–µ—Ç—Å—è, —Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ NO.

- –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –≤ –ø–µ—Ä–≤—ã—Ö –¥–≤—É—Ö —Å—Ç—Ä–æ–∫–∞—Ö –ø–æ–¥–∞—é—Ç—Å—è —á–∏—Å–ª–∞ 
m –∏ 
n ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –î–∞–ª–µ–µ –∏–¥—É—Ç 
m+n —Å—Ç—Ä–æ–∫ ‚Äî —Ñ–∞–º–∏–ª–∏–∏ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É, –≤ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

- –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∏–∑—É—á–∞—é—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö —É—á–µ–Ω–∏–∫–æ–≤ –Ω–µ –æ–∫–∞–∂–µ—Ç—Å—è, —Ç–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ NO.
